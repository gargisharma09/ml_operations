```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_excel(r"C:\Users\GARGI SHARMA\OneDrive\Desktop\Book1.xlsx")
df.head()
```

```python
# Step 2: Prepare the data
X = df[['square_feet','no of rooms','year']]  # Features (independent variable)
y = df['price']          # Target (dependent variable)
```

```python
# Step 3: Train the linear regression model
model = LinearRegression()
model.fit(X, y)
```

```python
# Step 4: Make predictions
predictions = model.predict(X)
```

```python
rmse = np.sqrt(mean_squared_error(y, predictions))
r2 = r2_score(y, predictions)

print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R-squared (R²): {r2}")
```

```python
plt.scatter(y, predictions, color='blue',label='Actual price')
plt.plot([min(y), max(y)], [min(y), max(y)], color='red', label='Prediction Line')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs Predicted Prices')
plt.legend()
plt.show()
```

```python
# Display the coefficient and intercept of the regression line
print(f"Coefficient: {model.coef_}")
print(f"Intercept: {model.intercept_}")
```

```python
example_square_feet = 1800
example_no_of_rooms = 0
example_year = 2000
predicted_price = model.predict([[example_square_feet, example_no_of_rooms, example_year]])
print('Predicted price for', example_square_feet, 'square feet:', predicted_price)
```
