### **outlier treatment** ###


```python
for i in xtrain.select_dtypes(exclude="object"):
    q1=data[i].quantile(0.25)
    q3=data[i].quantile(0.75)
    iqr=q3-q1
    u=q3+1.5*iqr
    l=q1-1.5*iqr
    data[i]=data[i].clip(lower=l,upper=u)
```

```python
for i in xtrain.select_dtypes(include=np.number):
    q1=data[i].quantile(0.25)
    q3=data[i].quantile(0.75)
    iqr=q3-q1
    u=q3+1.5*iqr
    l=q1-1.5*iqr
    data[i]=data[i].clip(lower=l,upper=u)
```
